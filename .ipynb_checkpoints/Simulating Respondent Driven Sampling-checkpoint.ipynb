# Simulating Respondent Driven Sampling with known Trait Homophily

```python
import itertools
import random
import math
import networkx as nx
import statistics
import numpy as np
import pandas as pd
import collections
import matplotlib.pyplot as plt
from networkx.generators.classic import empty_graph, path_graph, complete_graph

from collections import defaultdict

__all__ = ['fast_gnp_random_graph']

%matplotlib inline
```

## Function to create initial graph with specified traits

### 1. Create undirected graph with n nodes and p probability of edge creation.

```python
def fast_gnp_random_graph(n, p, seed=None, directed=False):
    """Return a multi directed graph G_{n,p} (Erdős-Rényi graph, binomial graph).

    Parameters
    ----------
    n : int
        The number of nodes.
    p : float
        Probability for edge creation.
    seed : int, optional
        Seed for random number generator (default=None).
    directed : bool, optional (default=False)
        If True return a directed graph

    Notes
    -----
    The G_{n,p} graph algorithm chooses each of the [n(n-1)]/2
    (undirected) or n(n-1) (directed) possible edges with probability p.

    This algorithm is O(n+m) where m is the expected number of
    edges m=p*n*(n-1)/2.

    It should be faster than gnp_random_graph when p is small and
    the expected number of edges is small (sparse graph).

    See Also
    --------
    gnp_random_graph

    References
    ----------
    .. [1] Vladimir Batagelj and Ulrik Brandes,
       "Efficient generation of large random networks",
       Phys. Rev. E, 71, 036113, 2005.
    """
    G = empty_graph(n)
    G.name="fast_gnp_random_graph(%s,%s)"%(n,p)

    if not seed is None:
        random.seed(seed)

    if p <= 0 or p >= 1:
        return nx.gnp_random_graph(n,p,directed=directed)

    w = -1
    lp = math.log(1.0 - p)

    if directed:
        G = nx.MultiDiGraph(G)
        # Nodes in graph are from 0,n-1 (start with v as the first node index).
        v = 0
        while v < n:
            lr = math.log(1.0 - random.random())
            w = w + 1 + int(lr/lp)
            if v == w: # avoid self loops
                w = w + 1
            while  w >= n and v < n:
                w = w - n
                v = v + 1
                if v == w: # avoid self loops
                    w = w + 1
            if v < n:
                G.add_edge(v, w)
    else:
        # Nodes in graph are from 0,n-1 (start with v as the second node index).
        v = 1
        while v < n:
            lr = math.log(1.0 - random.random())
            w = w + 1 + int(lr/lp)
            while w >= v and v < n:
                w = w - v
                v = v + 1
            if v < n:
                G.add_edge(v, w)
    return G
#Source: https://networkx.github.io/documentation/networkx-1.9/_modules/networkx/generators/random_graphs.html
```

### 2. Randomly assign a trait to all of the nodes in the network with specified proportion. 

```python
def assign_trait(graph, trait_name, trait_proportion):
    '''Graph = networkx graph
       Proportion = float between 0.0 and 1.0 for the proportion with a given trait'''
    
    # list of length of nodes in Graph with key/value pairs for having trait or not at specified proportion
    trait_list = [{trait_name:1} for i in range(int(trait_proportion*len(graph)))]
    trait_list += [{trait_name :0} for i in range(int(len(graph) - trait_proportion*len(graph)))]
    random.shuffle(trait_list)
    nx.set_node_attributes(graph, dict(enumerate(trait_list)))
```

### Write the function.

```python
def make_initial_graph(nodes, p_edge_creation, trait_name, trait_proportion):
    G = fast_gnp_random_graph(nodes, p_edge_creation)
    result = assign_trait(G, trait_name, trait_proportion)
    return G
```

### Create graph with 1000 nodes and 0.2% edge creation. 65% of nodes with female attribute

```python
G = make_initial_graph(1000, 0.02, 'trait', 0.65)
```

### Create a seed sample from the population with a specified proportion of nodes with a given trait

```python
def sample_from_graph(graph, proportion, trait, size):
    '''Graph is the networkx graph you are sampling from, Proportion is a float between 0.0 and 1.0 for the intended
    proportion of the trait in the sample, Trait is a String and is the trait you are interested in, Size is an int for the 
    amount of nodes in sample.'''
    
    # Separate subgraphs; with and without trait
    subgraph_with_trait = graph.subgraph([key for key, value in nx.get_node_attributes(graph, trait).items() if value == 1])
    subgraph_no_trait = graph.subgraph([key for key, value in nx.get_node_attributes(graph, trait).items() if value == 0])

    # Sample from subgraphs @ given proportion
    sample_with_trait = graph.subgraph(random.sample(subgraph_with_trait.nodes, int(proportion*size)))
    sample_no_trait = graph.subgraph(random.sample(subgraph_no_trait.nodes, (size-int(proportion*size))))

    return nx.compose(sample_with_trait, sample_no_trait)
```

```python
seed_sample = sample_from_graph(G, 0.65, 'trait', 150)
```

### Functions needed to perform the chain sampling

```python
def trait_proportion(graph, trait):
    '''graph is the networkx graph, trait is a string for the trait of interest
       returns the percentage of nodes in the graph with specified trait (0-100).'''
    return list(nx.get_node_attributes(graph, trait).values()).count(1) \
             / len(list(nx.get_node_attributes(graph, trait).values()))*100

def num_contacts(graph, sample):
    '''Return dictionary with node id and that nodes degree in graph for each node in seed'''
    node_id = [node for (node, val) in sample.degree]
    degree = [graph.degree[node] for (node, val) in sample.degree]
    merge_dict = dict(zip(node_id, degree))
    return merge_dict

def list_of_contacts(graph, sample):
    '''Return a dictionary with node id and a list of that nodes contacts in graph for each node in seed'''
    node_id = [node for node in list(sample.nodes())]
    contacts = [list(graph.neighbors(node)) for node in list(sample.nodes())]
    merge_dict = dict(zip(node_id, contacts))
    return merge_dict
```

## Chain Sampling Function 1
### Inputs:
1. the original graph
2. your seed sample from the original graph
3. the number of waves of sampling you want to conduct

### This function assumes:
1. probability of a node participating given that node has the trait as P(participate|trait) = 0.47
2. P(participate|no trait) = 0.65
3. 80% Trait Homophily i.e P(refer within trait|trait) = 0.8 and P(refer within no trait|no trait) = 0.8
4. The initial random sample from the seed sample will be between 20% and 33% the size of the seed sample

```python
def chain_sample1(graph, seed, waves, current_sample= 0, checker = 1, results_df = None):
    ''' seed is the starting sample
        waves is number of waves for iterating
        This function runs chain sampling and updates table with results for each wave'''

    if checker > waves:
        return
    
    elif current_sample == 0: #to start the recursion, we need to take 
        
        #initialize a dataframe to record results
        results_df = pd.DataFrame([[0,trait_proportion(seed, 'trait')]], columns=['sampling_wave', 'percent_with_trait'])
        
        #start with a random sample of nodes from seed graph between 20% and 33% the size of the seed graph
        current_sample = graph.subgraph(random.sample(list(seed.nodes()), \
                        k = random.choice(range(int(len(seed.nodes())/5), int(len(seed.nodes())/3)))))
        
        #update results table with current wave and waves proportion
        results_df.loc[len(results_df)] = [checker, trait_proportion(current_sample, 'trait')]
              
        #recursively run sample
        chain_sample1(graph, seed, waves, current_sample, sum([1, checker]), results_df)
        
        return results_df
    
    else:
        #dictionary with key/value as node/list of contacts in G
        contacts_dict = list_of_contacts(graph, current_sample)

        # mike pence referral
        put=[]
        for i in range(len(list(nx.get_node_attributes(seed_sample, 'trait')))): # for each node in the seed sample:

            if list(nx.get_node_attributes(seed_sample, 'trait').values())[i] == 1: # if female then...
                if int(np.random.choice([0,1], 1, p = [0.53, 0.47])) == 1: # 47% chance node with trait provide referrals...
                    if int(np.random.choice([0,1], 1, p = [0.4, 0.6])) == 1: #60% chance mike pence
                        for j in range(len(list(G.neighbors(i)))):                
                            if nx.get_node_attributes(G, "trait")[list(G.neighbors(i))[j]] == 1: #both have trait, refer
                                put.append([list(G.neighbors(i))[j]])
                    else: # not mike pence
                        put.append([x for x in list(G.neighbors(i))]) #refer all contacts

            else: # if male then...
                if int(np.random.choice([0,1], 1, p = [0.35, 0.65])) == 1: # 65% chance no_trait provide referrals...
                    if int(np.random.choice([0,1], 1, p = [0.4, 0.6])) == 1: #60% chance mike pence...    
                        for j in range(len(list(G.neighbors(i)))):                
                            if nx.get_node_attributes(G, "trait")[list(G.neighbors(i))[j]] == 0: #if both no trait, refer
                                put.append([list(G.neighbors(i))[j]])
                        else: # not mike pence
                            put.append([x for x in list(G.neighbors(i))]) #refer all contacts

        # remove reduncancies and None values from list
        flat_list = [item for sublist in put for item in sublist]
        res = []
        for i in flat_list:
            if i not in res and i != None:
                res.append(i)

                
        # update current sample with new res list
        current_sample = nx.subgraph(graph, res)
        
        
        #update results table
        results_df.loc[len(results_df)] = [checker, trait_proportion(current_sample, 'trait')]
       
        #recursively run sample
        chain_sample1(graph, seed, waves, current_sample, sum([1, checker]), results_df)
```

## Chain Sampling Function 1
### Inputs:
1. the original graph
2. your seed sample from the original graph
3. the number of waves of sampling you want to conduct

### This function assumes:
1. probability of a node participating given that node has the trait as P(participate|trait) = 0.47
2. P(participate|no trait) = 0.65
3. 60% of Respondents exhibit 100% trait homophiliy (refer all contacts within trait, none who dont share trait), the remaining
40% of respondents exhibit 50% trait homophily (refer all contacts regardless of trait).  
4. The initial random sample from the seed sample will be between 20% and 33% the size of the seed sample

```python
def chain_sample2(graph, seed, waves, current_sample= 0, checker = 1, results_df = None):
    ''' seed is the starting sample
        waves is number of waves for iterating
        This function runs chain sampling and updates table with results for each wave'''

    if checker > waves:
        return
    
    elif current_sample == 0: #to start the recursion, we need to take 
        
        #initialize a dataframe to record results
        results_df = pd.DataFrame([[0,trait_proportion(seed, 'trait')]], columns=['sampling_wave', 'percent_with_trait'])
        
        #start with a random sample of nodes from seed graph between 20% and 33% the size of the seed graph
        current_sample = graph.subgraph(random.sample(list(seed.nodes()), \
                        k = random.choice(range(int(len(seed.nodes())/5), int(len(seed.nodes())/3)))))
        
        #update results table with current wave and waves proportion
        results_df.loc[len(results_df)] = [checker, trait_proportion(current_sample, 'trait')]
              
        #recursively run sample
        chain_sample2(graph, seed, waves, current_sample, sum([1, checker]), results_df)
        
        return results_df
    
    else:
        #dictionary with key/value as node/list of contacts in G
        contacts_dict = list_of_contacts(graph, current_sample)

        #for each node in new sample, 47% chance for women and 65% chance for men to refer contacts @ p=0.8 within gender 
        put=[]
        for i in range(len(list(nx.get_node_attributes(seed_sample, 'trait')))):
            if list(nx.get_node_attributes(seed_sample, 'trait').values())[i] == 1: # if has trait then...
                if int(np.random.choice([0,1], 1, p = [0.53, 0.47])) == 1: # 47% chance to provide referrals...
                    for j in range(len(list(graph.neighbors(i)))):
                        if nx.get_node_attributes(graph, "trait")[list(graph.neighbors(i))[j]] == 1: #if both have trait, 80% chance to add to list
                            put.append(list(np.random.choice([None, list(graph.neighbors(i))[j]], 1, p=[0.2, 0.8])))
                        else: #if no trait, 20% chance to add to list
                            put.append(list(np.random.choice([None, list(graph.neighbors(i))[j]], 1, p=[0.8, 0.2])))
            else: # if male then...
                if int(np.random.choice([0,1], 1, p = [0.35, 0.65])) == 1: # 65% chance men provide referrals...
                    for j in range(len(list(graph.neighbors(i)))): 
                        if nx.get_node_attributes(graph, "trait")[list(graph.neighbors(i))[j]] == 0: #if neither have trait, 80% chance to add to list
                            put.append(list(np.random.choice([None, list(graph.neighbors(i))[j]], 1, p=[0.2, 0.8])))
                        else: #if trait, 20% chance to add to list
                            put.append(list(np.random.choice([None, list(graph.neighbors(i))[j]], 1, p=[0.8, 0.2])))

        # remove reduncancies and None values from list
        flat_list = [item for sublist in put for item in sublist]
        res = []
        for i in flat_list:
            if i not in res and i != None:
                res.append(i)

                
        # update current sample with new res list
        current_sample = nx.subgraph(graph, res)
        
        
        #update results table
        results_df.loc[len(results_df)] = [checker, trait_proportion(current_sample, 'trait')]
       
        #recursively run sample
        chain_sample2(graph, seed, waves, current_sample, sum([1, checker]), results_df)
```

```python
def graph_sampling(graph, seed, chain_sample):
    '''chain_sample is the dataframe returned from running the chain_sample function
    This will plot the percent of female in each wave of sampling and the baseline proportion
    from the original network.'''
    
    graph_proportion = trait_proportion(graph, 'trait')
    seed_proportion = trait_proportion(seed, 'trait')
    
    fig = chain_sample.plot(x='sampling_wave', figsize=(12,8), yticks=np.arange(0,110,10), lw=4, title= 'Percent With Trait for Each Wave of Chain Sampling'.format(seed_proportion))
    fig = plt.axhline(y=graph_proportion, color='black', linestyle='-', lw=2)
    return fig
```

```python
fig1 = graph_sampling(G, seed_sample, results1)
```
